Reduce the length of the noraml vector to 1 such that it forms the unit vector in the direction normal to the polane $H$

$$
n = \frac{w}{|w|}

Next consider a vector $A$ starting from any arbitary point $p(x_1,y_1,z_1)$ on the plane $H$ pointing towards $z$
$$
A = p - z = (x_1-x_0,y_1-y_0,z_1-z_0)
$$

we project the vector $A$ to the normal vector $n$ by taking dot product of $A$ and the normal vector $n$, which will give out the distance $D$ between the point $z$ to on a line perpendicular to plane $H$ as we are finding
$$
D = |A \cdot n| 
$$
$$
=  \frac{|(p-z) \cdot w|}{|w|}
$$
$$

$$
D = \frac{|-b - w^Tz|}{|w|}

$$
= \frac{|b + w^Tz|}{|w|}
$$
